```python
import os

def list_files(startpath):
    for root, dirs, files in os.walk(startpath):
        level = root.replace(startpath, '').count(os.sep)
        indent = ' ' * 4 * (level)
        print(f"{indent}{os.path.basename(root)}/")
        subindent = ' ' * 4 * (level + 1)
        for f in files:
            print(f"{subindent}{f}")

list_files("c:/users/sven/Documents/BEP/data")
```

```python
import os

def list_files(startpath):
    num_files = 0
    for root, dirs, files in os.walk(startpath):
        num_files += len(files)
    return num_files

dir_path = "c:/users/sven/Documents/BEP/data"

# Get the number of files in the directory
num_files = list_files(dir_path)

# Check if there are exactly 2 files in the directory
if num_files == 2:
    # Do your action here
    print("There are 2 files in the directory")
else:
    print(f"There are {num_files} files in the directory, but we expected exactly 2")
```
