```python
import matplotlib.pyplot as plt

# Sample data
data = {
    '2D': [0.53,0.52,0.51,0.56,0.60,0.54,0.54],
    '2D+': [0.62,0.74,0.75,0.77,0.63,0.70,0.72],
    '2D+ VI': [0.66,0.82,0.79,0.76,0.69,0.69,0.70]
}

# Boxplot settings
labels = ['2D', '2D+', '2D+ VI']
ylim = [0.45, 0.85]
ylabel = 'AUC'

# Create the boxplot
fig, ax = plt.subplots()
ax.boxplot(data.values(), labels=labels)

# Set y-axis limits and label
ax.set_ylim(ylim)
ax.set_ylabel(ylabel)
ax.set_xlabel('Input method')

# Show the plot
plt.savefig("Results_07062023.pdf")
plt.show()
```
