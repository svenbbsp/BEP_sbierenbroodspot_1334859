```python
import matplotlib.pyplot as plt
import csv
```

```python


# Initialize the data dictionary
data = {
    '2D': [],
    '2D+': [],
    '2D+ VI': []
}

# Open the CSV file
with open("C:/Users/Sven/Documents/BEP_sbierenbroodspot_1334859/Technical_research/AUC_results_csv.csv", 'r', encoding='utf-8-sig') as csvfile:
    reader = csv.reader(csvfile)

    # Iterate over each row in the CSV file
    for row_number, row in enumerate(reader):

        # Convert the row elements to numbers
        numbers = []
        for num in row:
            if num.strip():  # Check if the string is not empty
                numbers.append(float(num))

        # Append the numbers to the appropriate array in the data dictionary
        if row_number == 0 or row_number == 21 or row_number == 42 or row_number == 63:
            data['2D+'].extend(numbers)
        elif row_number == 64:
            data['2D'].extend(numbers)
        else:
            data['2D+ VI'].extend(numbers)

# Print the resulting data dictionary
print(data)
```

```python


# Boxplot settings
labels = ['2D', '2D+', '2D+ VI']
ylim = [0.30,0.85]
ylabel = 'AUC'

# Create the boxplot
fig, ax = plt.subplots()
ax.boxplot(data.values(), labels=labels)

# Set y-axis limits and label
ax.set_ylim(ylim)
ax.set_ylabel(ylabel)
ax.set_xlabel('Input method')

# Show the plot
plt.savefig("Results_12072023.pdf")
plt.show()
```
