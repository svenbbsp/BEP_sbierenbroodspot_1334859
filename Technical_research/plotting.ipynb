```python
import matplotlib.pyplot as plt
import numpy as np

# create some sample data
data = np.random.randint(0, 10, size=(5, 7))

# set up the figure and axes
fig, ax = plt.subplots()

# set the width of each bar
width = 0.1

# create a bar for each group
for i in range(5):
    # calculate the x position of the bars
    x = np.arange(7) + i * width
    # create the bars
    ax.bar(x, data[i], width=width, label=f'Group {i+1}')

# add labels and legend
ax.set_xlabel('Entries')
ax.set_ylabel('Values')
ax.set_xticks(np.arange(7) + width * 2)
ax.set_xticklabels([f'Entry {i+1}' for i in range(7)])
ax.legend()

# show the plot
plt.show()
```
