```python
from sklearn.manifold import TSNE
import pandas as pd
import matplotlib.pyplot as plt
import re
import csv
import numpy as np
```

```python
df = pd.read_csv('C:/Users/Sven/Documents/BEP_sbierenbroodspot_1334859/Pyradiomics/v5/output2.csv')
```

```python
df_zonder_na = df.dropna()
y = df_zonder_na['ID'].values
labels = [re.sub(r'\d+', '', s) for s in y]
# Define color code mappings
color_map = {'B': '#0000FF', 'M': '#FFA500'}

# Replace values with color codes
y = [color_map[i] for i in labels]
```

```python


df_without_diagnostics = df.drop(df.columns[:25], axis=1).dropna()
df_without_diagnostics
```

```python
X = df_without_diagnostics.values
# Create an instance of the TSNE class
tsne = TSNE(n_components=1, random_state=42)

# Apply t-SNE to the data
X_tsne = tsne.fit_transform(X)


plt.plot(X_tsne[:, 0], c=y)
plt.show()
```

```python
output = []
for i in range(len(X_tsne)):
    output.append(np.append(X_tsne[i],labels[i]))
```

```python
with open('tsne.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['x1', 'Label'])
    for row in output:
        writer.writerow(row)
```
