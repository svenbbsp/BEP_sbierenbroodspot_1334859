```python
import matplotlib.pyplot as plt
import numpy as np

# Define the data
data = np.array([[0.563, 0.6621, 0.711, 0.7568], [0.4774, 0.5435, 0.7201, 0.6273], [0.7219, 0.6608, 0.6939, 0.6642]])
x_labels = ['SVM', 'KNN', 'OE', 'NN']
y_labels = ['Naive', 'PCA', 'ANOVA']

# Define the bar width
bar_width = 0.2


# Define the x-axis values for the bars
x1 = np.arange(len(x_labels))
x2 = [x + bar_width for x in x1]
x3 = [x + bar_width for x in x2]

colors = ['#FFFF00', '#9ACD31', '#008000']

# Create the bar chart
fig, ax = plt.subplots()
ax.bar(x1, data[0], width=bar_width, label=y_labels[0], color=colors[0])
ax.bar(x2, data[1], width=bar_width, label=y_labels[1], color=colors[1])
ax.bar(x3, data[2], width=bar_width, label=y_labels[2], color=colors[2])
ax.set_ylim(0, 1)
ax.legend()

# Set the title and labels for the axes
#ax.set_title('Grouped Bar Chart with 2D Data')
ax.set_xlabel('ML Classifier')
ax.set_ylabel('AUC')

# Set the x-axis ticks and labels
ax.set_xticks(x2)
ax.set_xticklabels(x_labels)

# Show the chart
plt.show()
```
